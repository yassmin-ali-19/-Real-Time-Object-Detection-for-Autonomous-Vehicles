```python
import yaml


data = {
    'train': 'C:/Users/Raneem/Desktop/project car/train_balanced/images',
    'val': 'C:/Users/Raneem/Desktop/project car/val/images', 
    'test': 'C:/Users/Raneem/Desktop/project car/test/images',  
    'nc': 4,
    'names': ['car', 'traffic light', 'person', 'traffic sign']
}


with open('C:/Users/Raneem/Desktop/project car/Data.yaml', 'w') as file:
    yaml.dump(data, file, sort_keys=False)
```

```python
from ultralytics import YOLO
model = YOLO('yolo11m.pt')  
results = model.train(
    data=r"C:\Users\Raneem\Desktop\project car\Data.yaml",  
    epochs=60,             
    imgsz=640,             
    batch=8,               
    lr0=0.001,             
    optimizer='SGD',       
    weight_decay=0.0005,   
    patience=20,                        
    cos_lr=True,           
    label_smoothing=0.1, 
    augment=True,           
    project=r"C:\Users\Raneem\Desktop\project_results",  
    name="finetune_medium_reduced",  
    workers=2,
    device= 0, 
    save = True)
```
