```python
from ultralytics import YOLO
```

```python
 model = YOLO("yolo11m.pt")

results = model.train(
    data=r"/content/drive/MyDrive/kaggle_dataset/dataset.yaml",
    epochs=100,              # more epochs for large, unbalanced datasets
    imgsz=640,               # or 800 if GPU allows
    batch=16,                # depending on GPU memory (8‚Äì32 works)
    lr0=0.001,               # good starting LR
    optimizer='AdamW',       # usually better for large datasets
    weight_decay=0.0005,
    patience=20,             # early stopping if no improvement
    cos_lr=True,             # cosine learning rate schedule
    label_smoothing=0.1,
    augment=True,            # strong augmentation helps minor classes
    mixup=0.2,               # optional, blends images to regularize
    mosaic=1.0,              # keep mosaic augmentation enabled
    project=r"/content/drive/MyDrive/results",
    name="autonomous_yolo11m",
    workers=6,
    save=True,
    amp=True,

)
```
