```python
yaml_content = """
train: /content/drive/MyDrive/kaggle_dataset/train/images
val: /content/drive/MyDrive/kaggle_dataset/val/images

nc: 10
names: ['person', 'car', 'bus', 'truck', 'bike', 'traffic light', 'traffic sign', 'motor', 'rider', 'train']
"""


with open("/content/drive/MyDrive/kaggle_dataset/dataset.yaml", "w") as f:
    f.write(yaml_content)

print("dataset.yaml created successfully!")
```
